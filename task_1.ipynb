## Merge all customer information from three tables into one table
Both datasets contain the same values about customers, so it is advisable to merge them into one table for our analysis

```python
customer_demographic['customer_id'].iloc[-1]
## specifies the position of last row
```

We need to add 'Customer Id' column to new_customer_lists to enable us merge the tables vertically

```python
new_cust_lists.head()
```

```python
new_cust_lists.insert(0, 'customer_id', range(4001, 4001 + len(new_cust_lists)))
```

```python
new_cust_lists.head()
```

```python

# Merge dataframes using the customer_id column
customer_demographic = pd.merge(customer_demographic, cus_add, how='left', on='customer_id')
```

```python
customer_demographic.head()
```

```python
customer_demographic = customer_demographic.drop(['address','postcode','state','country','property_valuation'], axis=1)
```

```python
customer_demographic.head()
```

```python
new_cust_lists.head()
```

```python
new_df = pd.concat([customer_demographic, new_cust_lists], ignore_index=True, sort=False)
new_df.head(20)
```
